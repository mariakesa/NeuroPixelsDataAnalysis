```python
from brian2 import *
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
sv_path='/media/maria/DATA1/Documents/NeuroMatchAcademy2020_dat/all_spikes.npy'
```

```python
np.load(sv_path,allow_pickle=True)
```

```python
def for_spike_generator(spike_time_cells):
    spikes=[]
    cell_identity=[]
    for cell_ind in range(spike_time_cells.shape[0]):
        spikes_=[]
        for spk in list(spike_time_cells[cell_ind].flatten()):
            spikes_.append(np.round_(spk,3)*1000*ms)
        spikes_=list(np.unique(spikes_))
        spikes=spikes+spikes_
        cell_identity+=[cell_ind]*len(spikes_)
    inds=np.argsort(spikes)
    spikes=np.array(spikes)[inds]
    cell_identity=np.array(cell_identity)[inds]
    return spikes, cell_identity

def proc2(sv_path):
    spike_time_cells=np.load(sv_path,allow_pickle=True)[:1000]/1000
    spikes, cell_identity=for_spike_generator(spike_time_cells)
    np.save('/media/maria/DATA1/Documents/NeuroMatchAcademy2020_dat/spk_arr2.npy',spikes)
    np.save('/media/maria/DATA1/Documents/NeuroMatchAcademy2020_dat/cell_identity2.npy',cell_identity)
```

```python
proc2(sv_path)
```
